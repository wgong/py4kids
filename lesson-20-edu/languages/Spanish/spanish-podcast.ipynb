```python
import re
import requests
from bs4 import BeautifulSoup
from IPython.display import display, Markdown, Latex
```

```python
import pandas as pd 
```

```python
BROWSER_HEADERS = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"}
```

```python
URL = "https://podcasters.spotify.com/pod/show/danny-evans/episodes/The-Language-Tutor-Spanish---Episode-24---Stem-Changing-Verbs-1-e6epae"
```

```python
page = requests.get(URL, headers=BROWSER_HEADERS)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

```python
results = soup.findAll("a", class_="sc-hcLkzq cKyWTB")
print(len(results))
```

```python
print(results[0].prettify())
```

```python
BASE_URL = "https://podcasters.spotify.com/pod/show"
```

```python
item = results[0]
```

```python
item
```

```python
url = f"{BASE_URL}{item['href']}"
url
```

```python
title = item.find("img", class_="sc-jGNbpL jdUnzH")["alt"]
```

```python
title
```

```python
data = []
for item in results:
    url = f"{BASE_URL}{item['href']}"
    title = item.find("img", class_="sc-jGNbpL jdUnzH")["alt"]
    data.append(dict(url=url, title=title))
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.to_csv("Lang-Tutor-Spanish-podcasts.csv")
```
