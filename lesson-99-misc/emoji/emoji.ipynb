There are several ways to get emojis in Python.

## Using Unicodes

To do this, you need to replace "+" with "000" from the Unicode. For example, "U+1F600" will become "U0001F600". Then, you need to prefix the Unicode with "\". For example, to get the emoji "ğŸ˜ƒ", you would use the following code:

```python
print("\U0001F600")
```

## Using CLDR names

Emojis also have CLDR short names, which can also be used to get the emoji in Python. To do this, you need to replace the spaces with underscores in the CLDR short name. For example, the CLDR short name for the emoji "ğŸ˜ƒ" is "face with smiling eyes". To get the emoji, you would use the following code:


```python
print("hello", "\N{winking face}", emoji.emojize(":duck:"), "\U0001F923")
```
